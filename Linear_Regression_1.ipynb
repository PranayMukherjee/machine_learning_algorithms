```python
# LINEAR REGRESSION USING GRADIENT DESCENT

from numpy import *
from sklearn.datasets import load_iris
iris=load_iris()
print(iris.DESCR)
random.shuffle(iris.data)
#print(iris.data)
x_train=iris.data[:100,:3]
y_train=iris.data[:100,3]
y_train.shape=(100,1)
print(y_train)
x_train=insert(x_train,0,1,axis=1)
print(x_train)


```

```python
theta=random.random((4,1)) #initializing parameters
print(theta)
rate=0.03 #learning rate
def model(array): #hypothesis function
    value=0
    for i in range(4):
        value=value+(array[i]*theta[i][0])
    return(value) 
def cost(theta):
    sum=0
    for i in range(100):
        sum=sum+0.5*((model(x_train[i])-y_train[i])**2)
    return(sum)
```

```python
for k in range(10):
    for i in range(100):    #  STOCHASTIC GRADIENT DESCENT
        hyp=model(x_train[i])
        for j in range(4):
            theta[j][0]=theta[j][0]-rate*(hyp-y_train[i][0])*x_train[i][j]
        print(cost(theta))    
        print("end of training",i) 
    print("-----------------------------")    
```
