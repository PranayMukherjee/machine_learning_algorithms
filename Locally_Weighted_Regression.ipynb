```python
#LOCALLY WEIGHTED REGRESSION USING IRIS DATASET

from numpy import *
from sklearn.datasets import load_iris
iris=load_iris()
#print(iris.DESCR)
random.shuffle(iris.data)
x_train=iris.data[:100,:3]
y_train=iris.data[:100,3]
x_test=iris.data[132,:3]
y_test=iris.data[132,3]
#print(x_test)
#print(y_test)
x_train=insert(x_train,0,1,axis=1)
x_test=insert(x_test,0,1)
#print(x_train)
#print(x_test)
```

```python
theta=random.random((4,1)) #defining hypothesis
print(theta)
def model(array):
    value=0
    for i in range(4):
        value=value+ theta[i][0]*array[i]
    return value
```

```python
weight=zeros(100) #calculating the weights of every training example w.r.t the test point
bandwidth=0.2
rate=0.03
for i in range(100):
    d=x_train[i]-x_test
    f=matmul(matrix(d),matrix(d).T)
    weight[i]=exp(-f/(2*(bandwidth**2)))
print(weight)    
    
def cost(theta): #cost function
    sum=0
    for i in range(100):
        sum=sum+weight[i]*((model(x_train[i])-y_train[i])**2)  
    return sum    
```

```python
for k in range(20):
    for i in range(100): #Stochastic Gradient Descent
        for j in range(4):
            theta[j][0]=theta[j][0]-(rate*2*weight[i]*(model(x_train[i])-y_train[i])*x_train[i][j])
        print(cost(theta))
        print("training end",i)
    print("---------------------")    
```
