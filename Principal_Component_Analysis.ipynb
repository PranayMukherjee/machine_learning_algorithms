```python
import numpy as np
from sklearn.datasets import load_boston
boston=load_boston()
print(boston.data.mean(axis=0))
print(boston.data.std(axis=0))
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(boston.data)
boston_1=scaler.transform(boston.data)
print(boston_1)
```

```python
import matplotlib.pyplot as plt
for i in range(13):
    plt.scatter(boston_1[:,i],boston.target,color='green')
    plt.xlabel(boston.feature_names[i])
    plt.ylabel('predicted price in 10000$')
    plt.show()
```

```python
value=np.zeros((13,13))
for i in range(len(boston_1)):
    c=(boston_1[i,:])[:,np.newaxis]
    value=value+boston_1[i,:]*c
#print(value)    
print(value.shape) 
principal_matrix=value/len(boston_1)
print(principal_matrix)    
```

```python
eigen_values=np.linalg.eig(principal_matrix)[0]
len(eigen_values)
print(eigen_values)
```

```python
eigen_vectors=np.linalg.eig(principal_matrix)[1]
len(eigen_vectors)
print(eigen_vectors)
```

```python
boston_pca=np.zeros((506,13))
for i in range(len(boston_1)):
    for j in range(len(eigen_vectors)):
        boston_pca[i,j]=np.dot(eigen_vectors[j],boston_1[i])
print(boston_pca)        
```

```python
for i in range(13):
    plt.scatter(boston_pca[:,i],boston.target,color='blue')
    plt.xlabel(boston.feature_names[i])
    plt.ylabel("the predicted price in 10000$")
    plt.show()
    
```
